```python
import os
import pandas as pd
def marge_table(concat_names=None,work_path='.',save_name='marge.xlsx'):
    # concat_names= ['订单编号', '平台类型', '店铺名称', '仓库名称']
    files = os.listdir(work_path)
    excel_files = [i for i in files if i.split(".")[-1] in ['xls','xlsx']]
    if concat_names is not None:
        excel_files = [i for i in excel_files if all([x in list(pd.read_excel(i).columns) for x in concat_names])]
    else:
        concat_names = list(pd.read_excel(excel_files[0]).columns)
    print(excel_files)
    if os.path.exists(save_name) is False:
        df = pd.DataFrame(columns=concat_names)
        df.to_excel(save_name,index=False)
    for file in excel_files:
        df = pd.read_excel(save_name)
        new_df = pd.read_excel(file)
        new_df = new_df[concat_names]
        df = df.append(new_df)
        df.to_excel(save_name,index=False)
marge_table()
```

```python
import heapq
from openpyxl import load_workbook
def save_top_sheet(file,col1,col2,save_num):
       df = pd.read_excel(file)
       file_name = 'new_' + file
       new_df= df.to_excel(file_name,index=False)
       df = pd.read_excel(file_name)
       all_selected = list(set(df[col1].values))
       selected_res_list = []
       for one in all_selected:
              any_one = df[df[col1] == one]
              print(one,any_one[col2].sum())
              selected_res_list.append({'name':one,'sum_num':any_one[col2].sum()})
       selected_max = heapq.nlargest(save_num,selected_res_list,key=lambda s: s['sum_num'])
       print(selected_max)
       top_data = [df[df[col1] ==i['name']].sort_values(by=col2, ascending=False) for i in selected_max]

       try:
              for index,data in zip([i['name'] + '【' + str(i['sum_num']) + '】' for i in selected_max],top_data):
                     with pd.ExcelWriter(file_name,engine='openpyxl',mode='a') as writer:
                            data.to_excel(writer, sheet_name = str(index),index=False)
       except (ValueError) as e:
              print('已经有了sheet名')
# marge_table()
save_top_sheet('test.xlsx','产品名称','订单数量(基本)',20)
```
